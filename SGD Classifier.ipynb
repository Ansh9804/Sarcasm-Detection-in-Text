```python
import pandas as pd
import re
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import SGDClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report, accuracy_score
```

```python
df = pd.read_csv(r"D:\Downloads\nlp Dataset\train.csv")
```

```python
def clean_text(text):
    text = re.sub(r"http\S+|@\S+|#\S+", "", text)
    text = re.sub(r"[^a-zA-Z\s]", "", text)
    return text.lower().strip()

df['cleaned_tweets'] = df['tweets'].apply(clean_text)
```

```python
label_encoder = LabelEncoder()
df['label'] = label_encoder.fit_transform(df['class'])
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    df['cleaned_tweets'], df['label'], test_size=0.2, random_state=42
)
```

```python
vectorizer = TfidfVectorizer(max_features=5000)
X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)
```

```python
sgd_model = SGDClassifier(loss='log_loss', max_iter=1000, random_state=42)
sgd_model.fit(X_train_tfidf, y_train)
```

```python
sgd_preds = sgd_model.predict(X_test_tfidf)
print("SGD Classifier Accuracy:", accuracy_score(y_test, sgd_preds))
print(classification_report(y_test, sgd_preds, target_names=label_encoder.classes_))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix

# Confusion Matrix for SGD Classifier
sgd_cm = confusion_matrix(y_test, sgd_preds)

# Plot
plt.figure(figsize=(5, 4))
sns.heatmap(sgd_cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=label_encoder.classes_,
            yticklabels=label_encoder.classes_)
plt.title("SGD Classifier - Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.preprocessing import label_binarize
from sklearn.metrics import roc_curve, auc, precision_recall_curve, average_precision_score
import numpy as np

# Auto-detect and binarize labels
classes = np.unique(y_test)
y_test_bin = label_binarize(y_test, classes=classes)

# Predict probabilities
sgd_probs = sgd_model.predict_proba(X_test_tfidf)

# Number of classes
n_classes = len(classes)

# --- ROC Curve for each class ---
plt.figure(figsize=(8, 6))
for i in range(n_classes):
    fpr, tpr, _ = roc_curve(y_test_bin[:, i], sgd_probs[:, i])
    roc_auc = auc(fpr, tpr)
    plt.plot(fpr, tpr, label=f"Class {classes[i]} (AUC = {roc_auc:.2f})")
    
plt.plot([0, 1], [0, 1], 'k--', label="Chance")
plt.title("SGD Classifier - ROC Curve (One-vs-Rest)")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# --- Precision-Recall Curve for each class ---
plt.figure(figsize=(8, 6))
for i in range(n_classes):
    precision, recall, _ = precision_recall_curve(y_test_bin[:, i], sgd_probs[:, i])
    ap = average_precision_score(y_test_bin[:, i], sgd_probs[:, i])
    plt.plot(recall, precision, label=f"Class {classes[i]} (AP = {ap:.2f})")

plt.title("SGD Classifier - Precision-Recall Curve (One-vs-Rest)")
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
